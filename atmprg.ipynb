```python
class atm:
    def __init__(self):
        self.pin=''
        self.balance=0
        self.attempt=3
        self.menu()

    def menu(self):
      user_input=int(input("""
      Welcome to BOB
      How can I help you 
      1.Press 1 for create pin
      2.Press 2 for change pin
      3.Press 3 for check balance
      4.Press 4 for withdraw
      5.Press 5 credited amount
      6.Press 6 for exit
      """))
        
      if user_input==1: 
         self.create_pin() 
      elif user_input==2: 
         self.change_pin() 
      elif user_input==3: 
         self.check_balance() 
      elif user_input==4: 
         self.withdraw() 
      elif user_input==5: 
         self.credite()
      else: 
         self.exit()

    def create_pin(self):
       userpin=int(input("Create a pin"))
       self.pin=userpin
       print("Successfully created")
       balance=int(input("Enter your balance"))
       self.balance=balance
       self.menu()
        
    def change_pin(self):
      self.attempts=3
      while self.attempt>0:
          userpin=int(input("Create a pin"))
          if userpin==self.pin:
             newpin=int(input("Enter your new pin"))
             self.pin=newpin
             print("Successfully change")
             break
          else:
             print("Incorrect pin") 
             self.attempt-=1
      else:
          print("Please try again later") 
      self.menu()

    def check_balance(self):
        self.attempts=3
        while self.attempts>0:
            userpin=int(input("Create a pin"))
            if userpin==self.pin:
                print("Your balance is:",self.balance) 
                break
            else:
                print("Incorrect pin")
                self.attempt-=1
        else:
            print("Please try again later")
        self.menu()

    def withdraw(self):
        self.attempts=3
        while self.attempt>0:
            userpin=int(input("Create a pin"))
            if userpin==self.pin:
                amount=int(input("Enter a amount"))
                self.balance-=amount
                print("Successfully debited and your available balance is ",self.balance)
                break
            else:
                print("Incorrect pin") 
                self.attempt-=1
        else:
            print("Please try again later")
        self.menu()

    def credite(self):
        self.attempts=3
        while self.attempt>0:
            userpin=int(input("Create a pin"))
            if userpin==self.pin:
                 amount=int(input("Enter a amount"))
                 self.balance+=amount
                 print("Successfully debited and your available balance is ",self.balance)
                 break
            else:
                print("Incorrect pin") 
                self.attempt-=1
        else:
            print("Please try again later")
        self.menu()

    def exit(self):
        print("Thank you")
        exit() 

a=atm()
       
```
